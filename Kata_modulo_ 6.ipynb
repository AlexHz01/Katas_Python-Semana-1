Ejercicio 2: Trabajando con datos de una lista

En nuestro ejemplo, queremos mostrar planetas más cerca y más lejos del sol de un planeta que el usuario ingresa por teclado.

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']
```

```python
# Solicitamos el nombre de un planeta
user_planet = input('Porfavor introdusca el nombre del planeta  (con la inicial del planeta)')
```

```python
# Busca el planeta en la lista
planet_index = planets.index(user_planet)
```

```python
# Muestra los planetas más cercanos al sol
print('Aqui estan los planetas mas cerca que ' + user_planet)
print(planets[0:planet_index])
```

```python
# Muestra los planetas más lejanos al sol
print('Aqui estan los planetas mas lejos que ' + user_planet)
print(planets[planet_index + 1:])
```
