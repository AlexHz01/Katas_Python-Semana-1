# Ejercicio 1: Creación de un bucle "while"

```python
##declarar las variables donde pidan el nombre los planetas y el otro almacena los datos ingresados 

new_planet = ''
planets = []
```

```python
## el ciclo while, en donde new planet sea diferente de "done" sera ingresando planeta y cuando sea igual dejara de pedir planetas

while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Enter a new planet ')
```

## Ejercicio 2: Creación de un ciclo "for"

```python
##declarar las variables donde pidan el nombre de  los planetas y el otro almacena los datos ingresados 
new_planet = ''
planets = []
```

```python
## ## el ciclo while, en donde new planet sea diferente de "done" sera ingresando planetas y cuando sea igual dejara de pedir planeta, el for agrega
# # a la variable planet los datos de planets para despues imprimir 

while new_planet.lower() != 'done':
    if new_planet:
        planets.append(new_planet)
    new_planet = input('Enter a new planet ')

for planet in planets:
    print(planet)
```
