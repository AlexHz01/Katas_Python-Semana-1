Ejercicio - Escribir declaraciones if, else, y elif

```python
a = 1000
b = 200
c = 270
if a > b:
    if b > c:
        print ("a es mayor que b y b es mayor que c")
    else: 
        print ("a es mayor que b y b menor que c")
elif a == b:
    print ("a es igual que b")
else:
    print ("a es menor que b")
```
